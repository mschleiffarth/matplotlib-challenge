```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_data, trial_data, how = "outer", on = "Mouse ID")

# Display the data table for preview
combined_df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

group_data = combined_df.groupby(["Drug", "Timepoint"])
mean_data = group_data["Tumor Volume (mm3)"].mean()

final_mean = pd.DataFrame({"Tumor Volume (mm3)" : mean_data})
final_mean.reset_index(inplace=True)
final_mean
#data = combined_df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]



# Convert to DataFrame

# Preview DataFrame
#data
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
group_sem_data = combined_df.groupby(["Drug", "Timepoint"])
sem_data = group_sem_data["Tumor Volume (mm3)"].sem()

final_sem = pd.DataFrame({"Tumor Volume (mm3)" : sem_data})
final_sem.reset_index(inplace=True)
final_sem
# Convert to DataFrame

# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames

reformat_v2 = final_mean.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)') 
reformat_sem = final_sem.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)') 
reformat_v2
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)

timepoint = [0,5,10,15,20,25,30,35,40,45]
Tumor_Volume_Capomulin = reformat_v2["Capomulin"]
Tumor_Volume_Infubinol = reformat_v2["Infubinol"]
Tumor_Volume_Ketapril = reformat_v2["Ketapril"]
Tumor_Volume_Placebo = reformat_v2["Placebo"]

Capomulin, = plt.plot(timepoint, Tumor_Volume_Capomulin, color="red", label="Capomulin")
Infubinol, = plt.plot(timepoint, Tumor_Volume_Infubinol, color="blue", label="Infubinol")
Ketapril, = plt.plot(timepoint, Tumor_Volume_Ketapril, color="green", label="Ketapril")
Placebo, = plt.plot(timepoint, Tumor_Volume_Placebo, color="black", label="Placebo")


plt.title("Tumor Response to Treatment")
plt.xlabel("Time(days)")
plt.ylabel("Tumor Volume (mm3)")
plt.xlim(0,50)
plt.ylim(35,75)
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid()


plt.errorbar(timepoint, Tumor_Volume_Capomulin, yerr= reformat_sem["Capomulin"], fmt="o", color="r", alpha=0.5)
plt.errorbar(timepoint, Tumor_Volume_Infubinol, yerr= reformat_sem["Infubinol"], fmt="^", color="b", alpha=0.5)
plt.errorbar(timepoint, Tumor_Volume_Ketapril, yerr= reformat_sem["Ketapril"], fmt="s", color="g", alpha=0.5)
plt.errorbar(timepoint, Tumor_Volume_Placebo, yerr= reformat_sem["Placebo"], fmt="D", color="black", alpha=0.5)
    
    
plt.show()
# Save the Figure

```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
group_metSite_data = combined_df.groupby(["Drug", "Timepoint"])
mean_metSite_data = group_metSite_data["Metastatic Sites"].mean()

final_mean_metSite = pd.DataFrame({"Metastatic Sites" : mean_metSite_data})
#final_mean_metSite.reset_index(inplace=True)
final_mean_metSite.head()

# Convert to DataFrame

# Preview DataFrame
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
group_sem_metSite_data = combined_df.groupby(["Drug", "Timepoint"])
sem_metSite_data = group_sem_metSite_data["Metastatic Sites"].sem()

final_sem_metSite = pd.DataFrame({"Metastatic Sites" : sem_metSite_data})
#final_sem_metSite.reset_index(inplace=True)
final_sem_metSite.head()
# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
final_sem_metSite.reset_index(inplace=True)
final_mean_metSite.reset_index(inplace=True)

reformat_metSite_mean = final_mean_metSite.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites') 
reformat_metSite_sem = final_sem_metSite.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites') 
reformat_metSite_mean
# Preview that Reformatting worked
```

```python
# Generate the Plot (with Error Bars)
timepoint = [0,5,10,15,20,25,30,35,40,45]
metSite_Capomulin = reformat_metSite_mean["Capomulin"]
metSite_Infubinol = reformat_metSite_mean["Infubinol"]
metSite_Ketapril = reformat_metSite_mean["Ketapril"]
metSite_Placebo = reformat_metSite_mean["Placebo"]

Capomulin, = plt.plot(timepoint, metSite_Capomulin, color="red", label="Capomulin")
Infubinol, = plt.plot(timepoint, metSite_Infubinol, color="blue", label="Infubinol")
Ketapril, = plt.plot(timepoint, metSite_Ketapril, color="green", label="Ketapril")
Placebo, = plt.plot(timepoint, metSite_Placebo, color="black", label="Placebo")


plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.xlim(0,47)
plt.ylim(0,3.8)
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid()


plt.errorbar(timepoint, metSite_Capomulin, yerr= reformat_metSite_sem["Capomulin"], fmt="o", color="r", alpha=0.5)
plt.errorbar(timepoint, metSite_Infubinol, yerr= reformat_metSite_sem["Infubinol"], fmt="^", color="b", alpha=0.5)
plt.errorbar(timepoint, metSite_Ketapril, yerr= reformat_metSite_sem["Ketapril"], fmt="s", color="g", alpha=0.5)
plt.errorbar(timepoint, metSite_Placebo, yerr= reformat_metSite_sem["Placebo"], fmt="D", color="black", alpha=0.5)
    
    
plt.show()
# Save the Figure

# Show the Figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

mice = combined_df.groupby(["Drug", "Timepoint"])
final_mice = pd.DataFrame(mice.size().reset_index(name = "Mouse Count"))
final_mice.head()



# Convert to DataFrame

# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
final_mice_reformatted = final_mice.pivot(index='Timepoint', columns='Drug', values='Mouse Count')
# Preview the Data Frame
final_mice_reformatted.head()
```

```python
# Generate the Plot (Accounting for percentages)
timepoint = [0,5,10,15,20,25,30,35,40,45]
mouse_survival_Capomulin = final_mice_reformatted["Capomulin"]
mouse_survival_Infubinol = final_mice_reformatted["Infubinol"]
mouse_survival_Ketapril = final_mice_reformatted["Ketapril"]
mouse_survival_Placebo = final_mice_reformatted["Placebo"]

percentage_mouse_survival_Capomulin = (mouse_survival_Capomulin/25) * 100
percentage_mouse_survival_Infubinol = (mouse_survival_Infubinol/25) * 100
percentage_mouse_survival_Ketapril = (mouse_survival_Ketapril/25) * 100
percentage_mouse_survival_Placebo = (mouse_survival_Placebo/25) * 100

Capomulin, = plt.plot(timepoint, percentage_mouse_survival_Capomulin, color="red", label="Capomulin", marker="o")
Infubinol, = plt.plot(timepoint, percentage_mouse_survival_Infubinol, color="blue", label="Infubinol", marker="^")
Ketapril, = plt.plot(timepoint, percentage_mouse_survival_Ketapril, color="green", label="Ketapril", marker="s")
Placebo, = plt.plot(timepoint, percentage_mouse_survival_Placebo, color="black", label="Placebo", marker="D")


plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.xlim(0,47)
plt.ylim(33,105)
plt.legend(handles=[Capomulin, Infubinol, Ketapril, Placebo], loc="best")
plt.grid()

    
plt.show()
# Save the Figure

# Show the Figure
```
